```julia
f    = ROOTFile(
    "/home/shoram/Work/PhD_Thesis/Job15/AbsoluteAngles/MomPosEneThetaPhi_1e6pf.root",
);
l1 = LazyTree(f, "tree", ["x1Emitted", "y1Emitted", "z1Emitted", "x1Escaped", "y1Escaped", "z1Escaped"])
l2 = LazyTree(f, "tree", ["x2Emitted", "y2Emitted", "z2Emitted", "x2Escaped", "y2Escaped", "z2Escaped"])
tree = DataFrame( 
    xEmitted = vcat(l1.x1Emitted, l2.x2Emitted), 
    yEmitted = vcat(l1.y1Emitted, l2.y2Emitted),
    zEmitted = vcat(l1.z1Emitted, l2.z2Emitted),
    xEscaped = vcat(l1.x1Escaped, l2.x2Escaped), 
    yEscaped = vcat(l1.y1Escaped, l2.y2Escaped),
    zEscaped = vcat(l1.z1Escaped, l2.z2Escaped),
)
```

```julia
@chain tree begin
    @subset! abs.(:xEscaped) .<= 0.137
    @transform! :d = @. sqrt((:xEscaped - :xEmitted)^2+(:yEscaped - :yEmitted)^2+(:zEscaped - :zEmitted)^2)
    @subset! :d .< 5 # 3d distance traveled, cutting out the outliers (above 5 mm)
end
```

```julia
hist1DDistance = StatsBase.fit(Histogram, tree.d, nbins = 500)
stephist(
    tree.d, 
    nbins = 500, 
    label = "", 
    ylims = (0, 1.1*(maximum(hist1DDistance.weights))), 
    xlims = (0,5), 
    lw = 4, 
    xlabel = "d [mm]", 
    ylabel = "counts [#/0.01mm]", 
    legend = :topright
)

mu = round(mean(tree.d), digits = 3)
st = round(std(tree.d), digits = 3 )

vline!( [mean(tree.d)], lw = 3, label = "mean" )
vline!( [median(tree.d)], lw = 3, label = "median" )
vspan!( 
    [mean(tree.d) - std(tree.d), mean(tree.d) + std(tree.d) ], 
    c     = :black, 
    alpha = 0.2, 
    label = "mean ± σ = $mu ± $st" 
)
```

```julia
@chain tree begin
    @transform! :dmin = 0.125 .- abs.(:xEmitted) # distance to closest wall
    @transform! :dmax = 2*0.125 .- :dmin         # distance to furthest wall
    @rtransform! :t = :xEmitted .* :xEscaped .> 0 ? :dmin : :dmax # horizontal distance traveled
end
```

```julia
function get_bin_center1(x, xMin, xMax, nBins)
    binWidth = (xMax - xMin) / nBins
    if x == 0.0
        return binWidth/2
    end
    return ceil(x/binWidth)*binWidth - binWidth/2
end
```

```julia
yMin = -2500
yMax = 2500
Δy = 100
nY = (yMax - yMin)/Δy

zMin = -1350
zMax = 1350
Δz = 100
nZ = (zMax - zMin)/Δz
```

```julia
@rtransform! tree :yzBins = (
    get_bin_center.(:yEscaped, yMin, yMax, Int(nY)),
    get_bin_center.(:zEscaped, zMin, zMax, Int(nZ))
)
```

```julia
gdf = @chain tree begin
    groupby(:yzBins, sort =:true)
    @combine( :d_mean_μm = round(mean(:d) .* 1000))
end
```

```julia
h2d = Hist2D( 
    (Float64[], Float64[]),
    (yMin+Δy:Δy:yMax-Δy, zMin+Δz:Δz:zMax-Δz)
)
```

```julia
for row in eachrow(gdf)
    push!(h2d, row.yzBins[1], row.yzBins[2],  row.d_mean_μm)
end
```

```julia
plot(h2d, 
    c = :turbo, 
    xlabel = "y [mm]; Δy = $Δy mm", 
    ylabel = "z [mm]; Δz = $Δz mm", 
    colorbar_title = "mean distance [μm]"
)
```

```julia
sdf = @chain tree begin
    @select(:d, :t)
    @transform :dt = :d ./ :t
    @subset! 1.0 .< :dt .< quantile( tree.d ./ tree.t, 0.99 )  # subset of 99% data statistics, cut-off outliers       
end;
```

```julia
histogram2d(
    sdf.t, 
    sdf.d, 
    xlabel = "t", 
    ylabel ="d", 
    c = :jet, 
    colorbar_scale = :log10, 
    ylims = (0, 0.5)
)
```

```julia
xCutoff = quantile( tree.d ./ tree.t, 0.99 ); # find where 99% stat lies
```

```julia
dts = filter!( 
    x -> 1.0 .< x .< xCutoff, 
    tree.d ./ tree.t 
); # array of d/t values that fall within range 1.0 < x < 99% stats
```

```julia
hist1DdtRatio = StatsBase.fit(Histogram, sdf.dt, nbins = 100)
stephist( 
    sdf.dt, 
    ylims = (0.9*minimum(hist1DdtRatio.weights), 1.1*maximum(hist1DdtRatio.weights)), 
    legend= :none, 
    yscale = :log10, 
    xlims = (1,61), 
    nbins = 100,
    xlabel ="d/t", 
    lw = 4,
    ylabel = "counts [#/0.6mm]",
    title = "d/t; ratio of 3d distance traveled (d) \nto distance traveled along x-axis (t)"
)
# those events where d/t < 1 are such where the electron escaped within the source foil (not through Mylar)
```

```julia
Δt = 0.01
lims = (-2,2) # h2d limits on x,y axes
nBins = 400 # nbins 
```

```julia

anim = @animate for t in 0:Δt:0.25-Δt
    t2 = round(t+Δt, digits = 5)
    newDf = @chain tree begin
        @subset t .< :t .< t2
        @transform :y = :yEmitted - :yEscaped
        @transform :z = :zEmitted - :zEscaped
        @transform :r = @. sqrt( :y^2 + :z^2 )
        @select(:y, :z, :t, :r)
    end    
    
    bw = lims[2]/nBins
    
    h2 = histogram2d(
            newDf.y, 
            newDf.z, 
            lims = lims, 
            nbins = (nBins,nBins),
            xlabel = "y [mm] ",
            ylabel = "z [mm] ",
            clims  = (1, 1e2) 
        )
    
    h1 = stephist(
            newDf.r,
            lw = 4, 
            ylims = (0, 16),
            xlims = (0, 2),
            nbins = nBins,
            xlabel = "r [mm]",
            label ="",
            norm = :pdf
        ) 
        
    vline!(
        [quantile(newDf.r, 0.90)], 
        label = "90%",
        c = :red
    )
    vline!(
        [quantile(newDf.r, 0.95)], 
        label = "95%",
        c = :green
    )
    vline!(
        [quantile(newDf.r, 0.99)], 
        label = "99%",
        c = :blue
    )

    plot(
        h2, 
        h1,
        size = (1200, 800),
        plot_title= "t ∈ ($t, $t2) mm;"
    )
end
```
